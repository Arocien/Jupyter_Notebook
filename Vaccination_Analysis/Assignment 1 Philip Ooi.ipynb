### Wrangling Data

#### Wrangling vaccination data

```python
country_vax = country_vaccinations.filter(["country","people_fully_vaccinated","daily_vaccinations","vaccines"])
#filtering only the columns of interest 
country_vax
```

```python
#extracting the countries of interest Indonesia, Malaysia, Singapore, Thailand, Philippines, and Australia
country_filter = ["Indonesia","Malaysia","Singapore","Thailand","Philippines","Australia"]
boolean_series = country_vax.country.isin(country_filter) #returns boolean true or false
filtered_df = country_vax[boolean_series] #filt based on boolean True or False 
filtered_df
```

#### Why a list was chosen
A list was chosen as the way of data representation for the filter. It is a mutable data type meaning that the items within 
can be changed and values can be added later in the code using append() or pop(). This data type of chosen because it is a scalable data type, more values can be added to increase the scope of study, while the contrary of narrowing your scope is also possible through the manipulation of the country_filter. This is why list was chosen as the preferred data type for the assignment. Also if the filters are defined in a list, the filt variable can be used at multiple positions where the filtering is necessary.

```python
#aggregate total vaccinations from daily vaccinations (add daily together and replace)
func = {"daily_vaccinations":"sum","people_fully_vaccinated":"max"}
df1_fin = filtered_df.groupby("country").agg(func)
df1_fin = df1_fin.reset_index()
df1_fin


```

```python
df1 = filtered_df.drop_duplicates(subset = "vaccines")
df2_fin = df1.drop(["people_fully_vaccinated","daily_vaccinations"],axis = 1) #NB axis = 1 (columns) Axis = 0 (Rows)
df2_fin
```

```python
df_vaxFinal = df1_fin.merge(df2_fin,how ='inner')
df_vaxFinal
```

<br>

For the final question, you will probably need the non-aggregated data from 
“Country-Vaccination.csv”. You are to extract the data that’s related only to 
Australia and then plot a line graph on the daily_vaccinations over time. Like 
earlier, you are to discard the original total_vaccinations and create the total 
vaccinations for each day using the cumulative sum of the daily vaccinations (up to 
that day). Again, plot a line graph to visualise the cumulative vaccinations over time. 
Explain in what circumstances would the first line graph be useful (if at all) and in 
what circumstances that the second (cumulative) line graph would be useful?

```python
country_vaccinations = pd.read_csv("data/Country-Vaccinations.csv")
country_vaccinations
```

```python
country_filter = ["Australia"]
boolean_series = country_vax.country.isin(country_filter) #returns boolean true or false
aus_df = country_vax[boolean_series] #filt based on boolean True or False 
aus_df.reset_index(drop = True)
```

#### Final data for plotting 

```python
#drop NaN lines
df1= aus_df.dropna()
vax_aus = df1.drop(["people_fully_vaccinated","vaccines","country"], axis = 1)
df_ausvax = vax_aus.reset_index(drop = True)
df_ausvax
```

```python
#cumulative sum
cum_vax = vax_aus.daily_vaccinations.cumsum()
cumulative_vax = cum_vax.reset_index(drop = True)
cumulative_vax #data of vaccines through accumulation 
```

### Graph for daily vaccinations

```python
df_ausvax.plot.line()
plt.xlabel("Days") #getting our xlabels
plt.ylabel("Daily Vaccinations")
plt.title('Number of daily vaccinations over time') #titling labels
```

#### Graph for cumulative vax

```python
cumulative_vax.plot.line()
plt.xlabel("Days") #getting our xlabels
plt.ylabel("Daily Vaccinations (in millions)")
plt.title('Number of daily vaccinations over time') #titling labels
```

### Analysis of graphs 

The first graph shows the amount of daily vaccinations over time while the secondary graph shows the cumulative vaccinations over time. The first graph could be possibly used to check efficiency of vaccine rollout and if the vaccinations per day is rising over time. The secondary graph is a lot more useful as it shows us the total number of vaccines administered to the general public over time. This allows us to predict the general trend of vaccinations over the coming days and how many vaccinations in total that Australia has administered.
